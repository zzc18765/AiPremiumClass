```python
# å¼ é‡çš„å±æ€§æè¿°äº†å¼ é‡çš„å½¢çŠ¶ã€æ•°æ®ç±»å‹å’Œå­˜å‚¨å®ƒä»¬çš„è®¾å¤‡ã€‚ä»¥å¯¹è±¡çš„â»†åº¦æ¥åˆ¤æ–­ï¼Œ
# å¼ é‡å¯ä»¥çœ‹åšæ˜¯å…·æœ‰ç‰¹å¾å’Œâ½…æ³•çš„å¯¹è±¡ã€‚
# https://pytorch.org/docs/stable/torch.html 
# è¿™ä¸ªé“¾æ¥â¾¥ å…¨â¾¯ä»‹ç»100 å¤šç§å¼ é‡è¿ç®—ï¼ŒåŒ…æ‹¬ç®—æœ¯ã€çº¿æ€§ä»£æ•°ã€çŸ©é˜µæ“ä½œï¼ˆè½¬ç½®ã€ç´¢å¼•ã€åˆ‡â½šï¼‰ã€é‡‡æ ·ç­‰ç­‰ã€‚

tensor = torch.rand(3,4)
tensor.shape
tensor.dtype
tensor.device
print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

```python
# æ£€æŸ¥pytorchæ˜¯å¦æ”¯æŒGPU
if torch.cuda.is_available():
    device = torch.device("cuda")
    tensor = tensor.to(device)
print(tensor)
print(tensor.device)

# macä¸Šæ²¡æœ‰GPUï¼Œä½¿ç”¨Mç³»åˆ—èŠ¯ç‰‡
if torch.backends.mps.is_available():
    device = torch.device("mps")
    tensor = tensor.to(device)

print(tensor)
print(tensor.device)
```

```python
# å¼ é‡çš„ç´¢å¼•å’Œåˆ‡â½šï¼š
tensor = torch.ones(4, 4)
print('First row: ', tensor[0])
print('First column: ', tensor[:, 0])
print('Last column:', tensor[..., -1])
tensor[:,1] = 0  #å¯¹å¼ é‡ä¸­æ‰€æœ‰è¡Œçš„ç¬¬1åˆ—æ“ä½œ
print(tensor)
```

```python
# å¯ä»¥ä½¿â½¤ torch.cat â½¤æ¥è¿æ¥æŒ‡å®šç»´åº¦çš„â¼€ç³»åˆ—å¼ é‡ã€‚å¦â¼€ä¸ªå’Œ torch.cat åŠŸèƒ½ç±»ä¼¼çš„å‡½æ•°
# æ˜¯torch.stack
# å‚æ•° dim æŒ‡å®šäº†æ²¿å“ªä¸ªç»´åº¦è¿›è¡Œæ‹¼æ¥ã€‚å¯¹äº dim=1ï¼š
# å¦‚æœ tensor æ˜¯ä¸€ä¸ªäºŒç»´å¼ é‡ï¼ˆçŸ©é˜µï¼‰ï¼Œdim=0 è¡¨ç¤ºæ²¿ç€è¡Œçš„æ–¹å‘ï¼ˆå‚ç›´æ–¹å‘ï¼‰æ‹¼æ¥ï¼Œè€Œ dim=1 è¡¨ç¤ºæ²¿ç€åˆ—çš„æ–¹å‘ï¼ˆæ°´å¹³æ–¹å‘ï¼‰æ‹¼æ¥ã€‚
# å…·ä½“è€Œè¨€ï¼Œdim=1 æ„å‘³ç€å°†å¤šä¸ªå¼ é‡åœ¨åˆ—æ–¹å‘ä¸Šæ‹¼æ¥åœ¨ä¸€èµ·ã€‚è¿™ä¼šå¢åŠ åˆ—çš„æ•°é‡ï¼Œè€Œè¡Œæ•°ä¿æŒä¸å˜ã€‚
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1 * 3)
print(t1.shape)
```

```python
import torch
tensor = torch.arange(1,10, dtype=torch.float32).reshape(3, 3)
#reshapeé‡æ–°å®šä¹‰å½¢çŠ¶

# è®¡ç®—ä¸¤ä¸ªå¼ é‡ä¹‹é—´çŸ©é˜µä¹˜æ³•çš„å‡ ç§æ–¹å¼ã€‚ y1, y2, y3 æœ€åçš„å€¼æ˜¯ä¸€æ ·çš„ dot
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)

print(y1)
print(y2)

y3 = torch.rand_like(tensor) #åˆå§‹åŒ–
torch.matmul(tensor, tensor.T, out=y3)
print(y3)


# è®¡ç®—å¼ é‡é€å…ƒç´ ç›¸ä¹˜çš„å‡ ç§æ–¹æ³•ã€‚ z1, z2, z3 æœ€åçš„å€¼æ˜¯ä¸€æ ·çš„ã€‚
z1 = tensor * tensor
z2 = tensor.mul(tensor)

z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)

print(z1)
print(z3)
```
